## Business Understanding

Die Preise für Immoblien steigen kontinuierlich an. Der Preis einer Immobilie setzt sich dabei aus verschiedenen Merkmale der Immobilien, etwa dem Baujahr oder der Lage, zusammen. Für potentielle Käufer von Immobilien ist es oft schwer nachvollziehbar, wie sich der Preis einer Immobilie zusammensetzt. Somit ist es hilfreich, ein Tool anzubieten, welches basierend auf vorhandenen Daten von Immobilienverkäufen den Preis einer Immobilie samt ihrer Spezifika vorhersagt, um somit potentielle Verkäufer eine Unterstützung bei der Entscheidungsfindung zu bieten. 

Das Ziel dieser Fallstudie ist es, ein geeignetes Modell für die Vohersage von Immobilienpreise zu finden. Dabei werden neben einzelnen Regressionsmodellen auch hybride Regressionsmodelle betrachtet. Mithilfe des Modelles mit der besten Prognosegüte können in einem späteren Schritt der Fallstudie zuverlässliche Preisvorhersagen für Immobilien getroffen werden. Das Modell wird dabei in ein Tool eingebunden, in welches Nutzer die Werte einer Immobilie eingeben können, um somit den Preis der Immobilie vorherzusagen.

## Data Preparation

In diesem Abschnitt wird der Datensatz bereinigt und für das Training der Modelle vorbereitet. 

```python
''' --------- Beschreibung ---------

Diese Funktion dient zur Bereinigung und Vorbereitung der Daten für das Trainieren der Modelle.

Schritte:

1. Entfernung aller Duplikate, sofern vorhanden
2. Das Attribut Ausbaustufe kann nicht genutzt werden, da es ein String ist.
    2.1 Entfernen des Suffix des Strings
    2.2 Ersetzen von ',' mit '.' um den Wert in ein Float umzuwandeln
    2.3 Umwandlung zu einem Float
3. Das Attribut Kellerhöhe besteht ebenfalls aus Strings, für das Training der Modelle werden die Kategorien durch Integer ersetzt.
4. Alle nan values von 'Besonderheiten' werden zu 'Keine Auskunft' ersetzt.
5. Sollten weitere nan-Werte in dem Datensatz vorhanden sein, werden diese entfernt, um einen sauberen und für das Training der Modelle geeigneten Datensatz zu generieren.

    ------------------------------- '''
def clean_data(df):
    
    df['A_Index'].drop_duplicates(inplace=True)

    df['Ausbaustufe'] = df['Ausbaustufe'].replace(to_replace=r'^([0-9](,5)?) (Ebene|Ebenen)$', value=r'\1' , regex=True).apply(lambda x: x.replace(',', '.')).astype(float)
    
    df = df.replace({'Kellerhoehe': {np.nan: 0, 'Sehr Schlecht': 150, 'Schlecht': 175, 'Durchschnitt': 200, 'Gut': 225, 'Sehr gut': 250}})
  
    df['Besonderheiten'].fillna('Keine Auskunft', inplace=True)

    df = df.dropna().reset_index(drop=True)

    return df
```

```python
properties_df = clean_data(properties_df)
```

In diesem Schritt werden mithilfe des OneHot Encodings die nominalskalierten Attribute "Besonderheiten" und "Lage" in numerische Attribute umgewandelt.  

```python

ohe = OneHotEncoder(handle_unknown='error', sparse_output=False)

ohe_df = pd.DataFrame(ohe.fit_transform(properties_df[['Besonderheiten', 'Lage']]), columns=ohe.get_feature_names_out())
properties_df = properties_df.join(ohe_df)
properties_df.drop(['Besonderheiten', 'Lage'], axis=1, inplace=True)
```

In diesem Schritt werden mithilfe des OrdinalEncodings die ordinalskalierten Attribute "Gesamtqual", "Gesamtzustand", und "Steigung" in numerische Attribute umgewandelt.

```python

gq_array = ['Schlecht', 'Durchschnitt', 'Gut', 'Sehr gut']
gz_array = ['Sehr Schlecht', 'Schlecht', 'Durchschnitt', 'Gut', 'Sehr gut']
s_array = ['Keine/Kaum', 'Mittel', 'Stark']

ordi = OrdinalEncoder(categories=[gq_array, gz_array, s_array])

properties_df[['Gesamtqual', 'Gesamtzustand', 'Steigung']] = ordi.fit_transform(properties_df[['Gesamtqual', 'Gesamtzustand', 'Steigung']])
```

## Modeling und Evaluation


In diesem Schritt findet das Training der Modelle und deren Evaluation statt. Dafür werden zunächst jene Attribute entfernt, die nur sehr geringe oder keine Auswirkungen auf den Verkaufspreis haben. Ebenfalls wird der Verkaufspreis entfernt, da dieser vorhergesagt werden soll. 

Zum Training und Validierung der Modelle werden die Trainingsdaten in einen Trainings- und Validierungsdatensatz mithilfe einer definierten Größe und des Global states aufgeteilt. 


```python
x = properties_df.drop(["A_Index", "Besonderheiten_Keine Auskunft", "QualInnenfarbe", "Z_Verkaufspreis"], axis=1)
y = properties_df["Z_Verkaufspreis"]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=global_random_state) 
```

### Funktion zur Evaluation der Modelle

Die nachfolgende Funktion dient zur Evaluation der Modelle. Ähnlich wie bei den vorgestellten Paper wird zur Evaluation als Messwert der Root Mean Square Error (RMSE) verwendet. Der RMSE ist ein Maß zur Beurteilung der Prognosegüte und gibt an, wie gut die Vorhersage (des Verkaufspreises) im Durschnitt von den tatsächlichen Werten abweicht. Dabei ist zu beachten, dass je größer der RMSE ist, desto schlechter ist die Anpassung des Modells. Das Modell mit dem niedrigsten RMSE ist daher am besten für die Vorhersage des Verkaufspreises geeignet. 

Die definierte Funktion berechnet den RMSE mithilfe der tatsächlichen Verkaufspreise und den vorhergesagten Verkaufspreise und gibt das Ergebnis zurück.

```python
def print_rmse(algorithm_name, y_test, y_pred):
    rmse = metrics.mean_squared_error(y_test, y_pred, squared=False)
    return[algorithm_name, rmse]
```

### Aufzählung der genutzten Modelle

Folgende Modelle werden für das Training und die Evaluation im Rahmen der Immobilienpreisvorhersage verwendet: 

* Linear Regression
* Random Forest Regressor
* Lasso
* Gradient Boosting Regressor
* Gradient Boosting Regressor + Random Forest Regressor
* Ridge + Lasso
* Linear Regression + Gradient Boosting Regressor
* Linear Regression + Random Forest Regressor
* ElasticNet + Gradient Boosting Regressor
* Lasso + Gradient Boosting Regressor


Nachfolgend werden alle aufgezählten Modelle mithilfe der Trainings- und Validierungsdaten trainiert. Nach dem Training wird der RMSE des Modells berechnet, wobei alle individuellen RMSE Werte in einem gesammelten Dataframe abgespeichert werden. In einem späteren Schritt findet mithilfe dieses Dataframes die graphische Evaluaierung der Ergebnisse statt. 

Zunächst werden die individuellen Modelle trainiert.

```python
evaluation_df = []

lr = LinearRegression()
gbr = GradientBoostingRegressor()
rfr = RandomForestRegressor()
ridge = Ridge()
lasso = Lasso()
elasticnet = ElasticNet()


lr.fit(x_train, y_train)
y_pred = lr.predict(x_test)
evaluation_df.append(print_rmse("Linear Regression", y_pred, y_test))

rfr.fit(x_train, y_train)
y_pred = rfr.predict(x_test)
evaluation_df.append(print_rmse("Random Forest Regressor", y_pred, y_test))

lasso.fit(x_train, y_train)
y_pred = lasso.predict(x_test)
evaluation_df.append(print_rmse("Lasso", y_pred, y_test))

gbr.fit(x_train, y_train)
y_pred = gbr.predict(x_test)
evaluation_df.append(print_rmse("Gradient Boosting Regressor", y_pred, y_test))
```

In der nächsten Zelle werden mithilfe des VotingRegressors ausgewählte hybride Modelle trainiert.  Der VotingRegressor ist ein Ensemble-Meta-Regressor, der jeweils mehrere Basisregressoren an den gesamten Datensatz anpasst. Anschließend werden die individuellen Vorhersagen der Basisregressoren gemittelt, um eine endgültige Vorhersage zu erstellen.

```python
rdg_ls = VotingRegressor(estimators=[('rdg', ridge), ('ls', lasso)])
rdg_ls.fit(x_train, y_train)
y_pred = rdg_ls.predict(x_test)
evaluation_df.append(print_rmse("Ridge + Lasso", y_pred, y_test))

lr_gbr = VotingRegressor(estimators=[('lr', lr), ('gbr', gbr)])
lr_gbr.fit(x_train, y_train)
y_pred = lr_gbr.predict(x_test)
evaluation_df.append(print_rmse("Linear Regression + Gradient Boosting Regressor", y_pred, y_test))

lr_rfr = VotingRegressor(estimators=[('lr', lr), ('gbr', rfr)])
lr_rfr.fit(x_train, y_train)
y_pred = lr_rfr.predict(x_test)
evaluation_df.append(print_rmse("Linear Regression + Random Forest Regressor", y_pred, y_test))

el_gbr = VotingRegressor(estimators=[('el', elasticnet), ('gbr', gbr)])
el_gbr.fit(x_train, y_train)
y_pred = el_gbr.predict(x_test)
evaluation_df.append(print_rmse("ElasticNet + Gradient Boosting Regressor", y_pred, y_test))

lasso_gbr = VotingRegressor(estimators=[('ls', lasso), ('gbr', gbr)])
lasso_gbr.fit(x_train, y_train)
y_pred = lasso_gbr.predict(x_test)
evaluation_df.append(print_rmse("Lasso + Gradient Boosting Regressor", y_pred, y_test))

gbr_rfr = VotingRegressor(estimators=[('gbr', gbr), ('rfr', rfr)])
gbr_rfr.fit(x_train, y_train)
y_pred = gbr_rfr.predict(x_test)
evaluation_df.append(print_rmse("Gradient Boosting Regressor + Random Forest Regressor", y_pred, y_test))
```

In dem nächsten Schritt wird das Dataframe mit den RMSE Werten der individuellen und hybriden Modelle aufsteigend sortiert und mithilfe eines Balkendiagramms visualisiert. 

```python
evaluation_df = sorted(evaluation_df, key=lambda x: x[1])

evaluation_df = pd.DataFrame(data=evaluation_df, columns=['Name des Algorithmus', 'RMSE'])

fig, ax = plt.subplots(figsize=(8, 8))
ax = sns.barplot(data=evaluation_df, y="Name des Algorithmus", x="RMSE").set(title="Vergleich des RMSE von jedem Algorithmus")
```

Wie in der Grafik zu sehen ist, erzielt das hybride Modell bestehend aus dem Gradient Boosting Regressor und dem Random Forest Regressor die beste Prognosegüte, da dieses Modell den geringsten RMSE Wert auffweist. Für dieses hybride Modelle werden alle Metriken ausgegeben. Mithilfe des hybriden Modells wird ein Score von über 89% erzielt.

```python
gbr_rfr = VotingRegressor(estimators=[('gbr', gbr), ('rfr', rfr)])
gbr_rfr.fit(x_train, y_train)
y_pred = gbr_rfr.predict(x_test)

print("Gradient Boosting Regressor + Random Forest Regressor" + ':')
print("   R2 / Score:", metrics.r2_score(y_test, y_pred))
print("   MSE:", metrics.mean_squared_error(y_test, y_pred))    
print("   RMSE:", metrics.mean_squared_error(y_test, y_pred, squared=False))
print("   MAPE:", metrics.mean_absolute_percentage_error(y_test, y_pred))
print("   MAX:", metrics.max_error(y_test, y_pred))
```
